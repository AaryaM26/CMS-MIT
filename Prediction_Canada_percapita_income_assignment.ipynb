Predict canada's per capita income in year 2020. Using this build a regression model and predict the per capita income of canadian citizens in
year 2020

```python
import pandas as pd
```

```python
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/drive/MyDrive/ML LAB/canada_per_capita_income.csv')
```

```python
df
```

```python
df.columns
```

```python
%matplotlib inline
plt.xlabel('year')
plt.ylabel('income')
plt.scatter(df['year'],df['per capita income (US$)'],color='red',marker='+')
```

```python
new_df = df.drop('per capita income (US$)',axis='columns')
new_df
```

```python
income = df['per capita income (US$)']
income
```

```python
reg = linear_model.LinearRegression()
reg.fit(new_df,income)
```

```python
reg.predict([[2020]])
```

```python
reg.coef_
```

```python
reg.intercept_
```
